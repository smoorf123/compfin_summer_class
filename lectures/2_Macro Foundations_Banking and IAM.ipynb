### Fractional Reserve Banking

>Fractional reserve banking is a system in which only a fraction of bank deposits are required to be available for withdrawal. Banks only need to keep a specific amount of cash on hand and can create loans from the money you deposit. Fractional reserves work to expand the economy by freeing capital for lending. Today, most economies' financial systems use fractional reserve banking.

Long story short, based on regulations, banks only keep a percentage of your money on hand any given moment. Through this they are able to use deposited capital towards lending operations.

Some key things to note:
- On March 26, 2020, the Federal Reserve reduced reserve requirements for all depositary institutions to zero. Instead, banks are now paid a specific interest rate on their reserve balance to encourage holding reserves.
- There are explicity limitations on what the banks can and cannot do with deposits through legislation. The Dodd-Frank Act of 2010 (post 2008 recession) with the Volcker Rule, Glass-Steagall from 1933, etc. to name a few key ones.
- To incentivize deposits, banks offer "interest" on any deposits you make with them. These range from near 0% (I'm looking at you Chase), to competitive, close to Fed Funds Rate interest rates (currently around 4% ish).

As this simple example illustrates, banks operate like any other business with assets, liabilities and equity. A very basic accounting lesson:

Now we go back to our initial example of banks and what their balance sheets look like. Here, Assets = Loans + Reserves + Short-term Treasuries; Liabilities = Deposits + Equity (Capital)



A well-functioning bank earns a **spread** between:
 - The interest on loans it issues, and
 - The interest it pays on deposits (savings/checking accounts, etc.)

E.g. A bank may issue a credit card at a 25% interest rate (the interest on a loan or credit card is also known as APR % which is the interest rate + any fees) and provides a 4.2% interest rate on a High-Yield Savings Account (aka HYSA) (the interest on a deposit is also known as APY %). As a simplified example, say the owner of the credit card borrowed \\$100 and missed the payment and now owes \\$25 in interest, simultaneously for someone depositing \\$100, they are now owed \\$4.20. The bank makes the difference i.e. \\$25 - \\$4.20 = \\$20.80. This \\$20.80 per \\$100 represents the bank’s **gross interest margin** and helps cover things like operating costs, default risk, regulatory capital requirements, and ultimately contributes to profits.

Note the word **spread**, it will appear often and just implies the gap or difference between two related values. 

We must also realize that this model works only as long as depositors trust that they can redeem their money on demand. That’s why liquidity and government regulation matter.

### Risks Associated

We briefly mentioned **Credit Risk** and **Interest Risk**. Let us define these two as they are incredibly important and will pop up often when discussing risks associated in finance going forward.

**Credit Risk** is essentially just the risk associated with a risky borrower and their inability to repay their debt. This is measured in many ways one of which is a **Credit Score** for ordinary individual debtors such as credit card holders.

**Interest Rate Risk** is a little more complex where it implies the risk of changing interest rates (Fed Funds Rate from previous lecture) and how that may affect bank interest margins mentioned above. 

There are two key types of loan structures in this context:
 - Fixed-rate loans, where the interest rate remains constant for the life of the loan
 - Variable-rate (or floating-rate) loans, where the rate adjusts periodically based on a benchmark (e.g., LIBOR, SOFR, or Fed Funds rate)

Suppose the Fed significantly reduces the Federal Funds Rate—from, say, 4% to 1%—in response to an economic downturn. This has two major effects:

 - Newly issued loans will carry much lower interest rates, reducing future interest income for banks.
 - Existing fixed-rate loans do not reprice, meaning banks are "locked in" to older, higher-rate lending—but this might initially look like an advantage.

However, the real issue arises when banks:
- Hold long-duration fixed-rate assets (e.g., 30-year mortgages issued at 6%)
- While funding them with shorter-duration or variable-rate liabilities (e.g., deposits whose rates quickly adjust downward)

In this case, if the cost of funding drops (due to lower deposit rates), the bank’s interest margin might improve temporarily.

But the reverse is more dangerous: when the Fed raises rates sharply (like in 2022–2023), banks with long-duration fixed-rate assets and shorter-term liabilities can see their interest margin shrink. This exact dynamic contributed to the collapse of Silicon Valley Bank, which had loaded up on long-term Treasuries and Mortgage-Backed Securities during the low-rate era and suffered large unrealized losses as rates rose.

## Types of Banking and Associated Regulation

### Retail/Commercial Banks

Retail banks, also known as commercial banks, are the most common type of banks. They serve individuals, small businesses, and large corporations. Retail banks provide services such as checking and savings accounts, loans, credit cards, mortgages, and other financial products such as CDs, mutual funds, and individual retirement accounts (IRAs). They may have physical branches, online banking platforms, and ATMs. 

The largest retail banks in the U.S. are JP Morgan Chase, Bank of America, Wells Fargo and Citigroup. Larger retail banks also cater to high-net-worth individuals with specialty services such as private banking and wealth management services.

#### Regulation: 
These banks are generally part of the Federal Reserve system and deposits may be insured by the FDIC. This is required in the case of nationally-chartered banks.

#### Risk Considerations:
- Interest Rate Risk: Especially during Fed tightening cycles.
- Liquidity Risk: Sudden deposit withdrawals (e.g., post-SVB).
- Credit Risk: Defaults from borrowers.

### Credit Unions

Credit unions are not-for-profit financial cooperatives owned and operated by their members. Members purchase shares in the co-op, and that money is pooled together to fund the credit union's loans. They serve specific communities, such as employees of a certain company, members of an organization, or residents of a particular geographic area, and are generally tax-exempt. Examples include Navy Federal Credit Union (for members and relatives of those in the Armed Forces, DOD, and Veterans), State Employees' Credit Union (SECU), etc.

Credit unions offer similar services to retail banks, including savings accounts, loans, and other financial products. They often provide more personalized service and may offer lower fees and higher interest rates on savings accounts. They tend to provide a limited range of services compared to banks. They also have fewer locations ATMs. 

#### Regulation: 
Credit Unions are not regulated/insured like commercial banks through FDIC/Federal Reserve System. The **National Credit Union Administration (NCUA)** is in charge of regulating Credit Unions at the federal level. All deposits at federally insured credit unions are protected by the **National Credit Union Share Insurance Fund (NCUSIF)**.

#### Risk Considerations:
- Interest Rate Risk: Similar to commercial banks. It can be said Credit Unions have more fixed-rate loan offerings, making them particularly vulnerable
- Liquidity Risk: Sudden deposit withdrawals (also lack same short-term liquidity funding options like discount window)
- Credit Risk: Similar to commercial banks.
- Concentration Risk: Members are often closely located geographically/occupationally thus leading to systemic risk possibilites.

### Investment Banks

Investment banks primarily cater to corporations, pension funds, governments, hedge funds, and other institutional clients rather than individual consumers. They specialize in providing financial advice, underwriting securities, facilitating mergers and acquisitions, and assisting with capital raising through stock and bond offerings. Investment banks play a crucial role in the capital markets and are involved in complex financial transactions. 

Morgan Stanley and Goldman Sachs are among the biggest U.S. investment banks.

#### Regulation: 
- Regulation of Investment banks is complex based on the operations of the particular bank. For broker-dealer activities, public securities issuance, disclosure, insider trading, and compliance with securities laws the **Securities and Exchange Commission (SEC) and Financial Industry Regulatory Authority (FINRA)** oversee operations. 
- If the investment bank is part of a bank holding company (e.g., Goldman Sachs post-2008), **the Fed** regulates capital adequacy, liquidity, stress testing (CCAR), and systemic risk.
- In the case of commodities and derivatives markets, regulation is done by **Commodity Futures Trading Commission (CFTC)**.

Similarly other federal agencies that may oversee operations include **Financial Stability Oversight Council (FSOC), Office of the Comptroller of the Currency (OCC), and the U.S. Treasury (through the IRS, OFAC)**.

#### Risk Considerations:
- Market Risk: Exposure to interest rate changes, credit spreads, equity prices, foreign exchange, and commodity prices.
- Credit Risk: Counterparty risk from trading partners, especially in derivatives and structured products (e.g. 2008 financial crisis; ABACUS Deal).
- Liquidity Risk: In the case of Bank Holding Companys (BHCs) part of the Fed, they rely heavily on short-term funding (e.g., repo markets).
- Systemic Risk: Large investment banks are interconnected with global financial markets. Not only do they provide crucial financial services such as underwriting and market making, they may also be a counterparty to billions of dollars worth of obligations potentially causing a cascading effect in the industry.

## Asset Management vs Banking

When we talk about large institutional holdings of money, it is easy to get confused between banks and asset managers. It is key to understand the differences to be able to talk about their roles in the world of finance.

Asset Managers:

- Do not take deposits
- Do not lend money from their own balance sheets
- Do not participate in the payment system (ACH, etc.)
- Are not subject to capital adequacy requirements like banks

Instead, they manage assets on behalf of clients, investing in stocks, bonds, alternatives, etc.

This does not mean they are unimportant in the world of finance, on the contrary, asset managers may be in control of Billions if not Trillions of dollars, giving them **massive influence** over markets.

### Key Types of Institutions and the Assets they Manage:
- **Mutual Funds & ETFs:** Pooled investment vehicles for retail and institutional investors e.g. Vanguard 500 Index Fund (VFIAX), SPDR S&P 500 ETF Trust (SPY).
- **Pension Funds:** Long-term retirement funds with strict fiduciary mandates e.g. Government Pension Fund of Norway ($1.9 Trillion as of June 2025). 
- **Endowments, Foundations and Sovereign Wealth Funds:** Often use multi-asset strategies with alternative investments e.g. Yale Univesity Endowment, Saudi Public Investment Fund (PIF).

Insurance companies can also be considered as non-traditional asset managers as they invest to match long-term liabilities, subject to asset-liability matching constraints. Insurance companies build diversified portfolios: Treasuries, corporates, mortgage-backed securities, private placements, sometimes even equities to match the duration and cash flow structure of expected claims. Examples of insurers can include Prudential, AIG, etc. Insurance companies may also outsource their asset management to firms such as BlackRock, etc.

### Active vs. Passive Management
It is important to note the difference, this will come up in later lectures as well as BUFN400.

 - **Active Management:** Attempts to outperform a benchmark (e.g., through stock picking or tactical asset allocation).
 - **Passive Management:** Tracks a market index (e.g., S&P 500). Significantly lower costs and is now predominantly the location of choice for most capital.

**Note:** BlackRock, Vanguard, and State Street collectively manage trillions of dollars and are integral to financial market functioning. Though not banks, they are now considered "too big to ignore" by regulators. This has raised corporate governance and federal regulation concerns, as a few institutions wield large voting blocks in the biggest public companies. This is a contentious topic however, especially from the perspective of these asset managers who say they exert minimal decision-making influence on the boards of companies they own shares of.

### Regulation:

Asset management firms fall under the category of non-bank financial institutions (NBFIs) or institutional investors.

More specifically:
- Regulated by the SEC under the Investment Advisers Act of 1940
- If offering mutual funds or ETFs, also regulated under the Investment Company Act of 1940
- Do not fall under the Federal Reserve or FDIC (unless affiliated with a bank)

## What about Hedge Funds and Private Equity? Are they an Institutional Asset Manager?

Hedge Funds and Private Equity are also major players in the world of finance and will round out this lecture as we understand how money is held and transacted in the world of finance. 

To answer the question, yes, these firms invest in non-traditional asset classes or use complex strategies, and can be considered **Alternative Asset Managers**. They often face fewer regulatory restrictions and are seen as (as the name suggests) alternatives for investing your cash besides the traditional equity offerings of the market.

### Hedge Funds
Hedge funds are typically structured as private investment partnerships and are not open to the general public. Access is usually limited to accredited investors or qualified purchasers, who meet certain financial requirements, signifying they can potentially handle the higher risks associated with hedge fund investments.

They use active, often high-risk strategies: long-short equity, macro trading, arbitrage, event-driven, etc. Their **legal structure** is of great importance, with the structuring usually involving Limited Partnerships (LPs) with a General Partner managing the fund. More on this on the Firm Structures lecture. Also to note is their manager compensation mechanisms ("2 and 20") which offer a unique risk-reward mechanism to their clients (highly researched topic; incredibly interesting). Regulation of Hedge Funds falls under the SEC.

Hedge Funds are **incredibly important** holding trillions of dollars of assets, with firms like Citadel and AQR, and are pioneers of industry when it comes to innovative strategy and generating income for their clients. Thus, they are studied for their outsized effect on markets as well as strategies for generating income ("alpha"), managing risk and complying with regulations. They will be discussed in length in BUFN400 as well as BUFN402. 

### Private Equity

Private equity (PE) is an alternative investment asset class that involves directly investing in private companies or conducting buyouts of public companies, resulting in the delisting of public equity. PE firms aim to improve the profitability and value of these companies with the ultimate goal of selling them for a profit. 

Some key strategies include:

- **Buyouts (including Leveraged Buyouts or LBOs):** Acquiring a majority or controlling stake in established, often mature companies, sometimes using significant debt to finance the acquisition (leveraged buyout or LBO).
- **Venture Capital (VC):** Investing in early-stage startups and emerging companies with high growth potential, usually taking a minority stake. VC is often considered a subset of private equity.
- **Growth Equity:** Providing capital to established and profitable companies seeking to expand, enter new markets, or improve operations.

Private Equity generally operates in 4 steps including but not limited to:
- Fundraising: PE firms raise capital from institutional investors (like pension funds, endowments) and high-net-worth individuals.
- Investment: The collected funds are used to acquire equity stakes in target companies.
- Value Creation: PE firms work with the management of the acquired companies to improve their performance, which may involve restructuring, implementing new strategies, or facilitating mergers and acquisitions.
- Exit: PE firms typically seek to exit their investments after a period, usually by selling their stake through an IPO, strategic acquisition, or secondary sale.

Private equity is also regulated by the SEC and they are also generally organized into Limited Partnerships (LPs).

>Relevant Example: The acquisition of Twitter by Elon Musk involved a significant private equity component, transforming the publicly traded company into a privately held one. Musk funded the $44 billion purchase using a mix of debt and equity, including investments from external parties and his own assets. This move eliminated Twitter's public reporting requirements and gave Musk greater control over the platform.

**Note:** The terms Hedge Funds and Private Equity are sometimes confused, even by the Media and professionals. ### Illiquidity and Volatility

Something we discussed in previous lectures is the concept of illiquidity and volatility of assets. This is especially true in the case of Hedge Funds and Private Equity as they take leveraged, directional, or short-term trading positions. Private Equity is worse as all operations happen in opaque markets and private valuations are the only way of finding out if any money was made (what incentives does this promote?). This leads to generally greater volatility than traditional forms of investment such as mutual funds or the SP500. 

In terms of illiquidity, in Private Equity, as part of the agreement, capital may be locked up for 7–10 years, and returns only materialize upon exit events (IPO, sales, etc.). For Hedge Funds, lock up periods may be 6 months to 2 years with a notice required for redemption (30-90 days). Underlying assets held by hedge funds may also not be easy or quick to sell e.g. distressed debt or credit arbitrage.
