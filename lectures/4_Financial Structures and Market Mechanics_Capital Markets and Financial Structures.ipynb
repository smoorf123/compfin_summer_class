## How does a company raise money? 

Traditionally when a company wants to raise money they have one of two options:

1) Equity Financing
2) Debt Financing

We discussed what the differences between equity and debt are last lecture and this builds upon those same concepts. Conceptually, this makes sense- if a company wants money they have to give something up. This can either be equity or debt (we will discuss the others later).

### Equity Financing

Equity financing is generally of three types. This type of funding the operations of the company (financing) is done through giving up *equity* or in other words ownership of the company and its profits.

#### (1) Initial Public Offering (IPO):
A company can raise capital by going from privately held to 'public', by selling shares to the public for the first time, raising significant capital and allowing retail and institutional investors to invest in the company, trading equity for capital. This involves listing it on a stock exchange (NYSE or NASDAQ in the U.S. generally). Previously, the company's ownership would have been held by private investors (founders, venture capitalists, etc.), however, after an IPO, ownership is divided among many more shareholders who buy shares on the public market.

IPOs have the potential to raise billions of dollars given the size of investors that are now able to buy shares in the company. Potential reasons for IPO*ing* may also include an exit strategy for the founders and early investors as well as increasing credibility through the transparency and scrutiny that comes with a public listing.

#### (2) Private Equity
Specialized firms raise capital from investors and pool these funds to invest in or acquire private companies, who exchange their equity for said capital. Their goal is to improve the value of these companies and then sell them for a profit, usually through a merger or acquisition or an initial public offering (IPO). These firms are referred to as private equity firms. 

Private equity is often seen as a very opaque industry given its functioning behind closed doors. One of if not the most important part of investing is being able to value your investment. This is easiest when you can find data on your company. Where public companies have to be entirely transparent and disclose all financials through 10-Ks etc., private companies do not do so. Thus, when private equity firms invest in private companies it involves a lot of subjective valuation away from the eyes of the public. This can cause many issues especially regarding trust in firms valuing their investments correctly.

#### (3) Venture Capital
Similar to private equity, but typically focuses on early-stage, high-growth companies. This is generally not a firm (with a manager making investment decisions on behalf of the pooled funds from clients), rather a couple of key wealthy investors who are willing to take the high risk involved with funding startups and early-stage companies. For the wealth they contribute, investors get compensated with equity. Sometimes companies may also give VCs (Venture Capitalists aka people engaging in venture capital) equity for their connections in the industry or the knowledge and experience they may have.

### Debt Financing

## Market Infrastructure and Different types of Investment Tools

Here we will break down the main three types of financial investment tools for better understanding the world of finance.

### Stocks
Everyone has an intuitive understanding of stocks but we can break this down further into its respective types. A simple refresher: a stock of a company is the contractual agreement between an investor and the company to a fractional ownership of the issuing company, in proportion to the amount of stock owned by the investor. When you buy a stock, you become a shareholder with a claim on the company’s assets and earnings.

#### Common Stock vs Preferred Stock
There are two main types of stock, namely common vs preferred stock. They both entitle an investor to the company's assets and earnings however there are significant differences between them. 

>A key difference is that preferred stock shareholders have priority over a company's income. They're paid dividends before common stock shareholders. Preferred stockholders are also paid first if a company is liquidated. Common stock shareholders are last in line when it comes to company assets. They're paid after creditors, bondholders, and preferred stock shareholders.

>Another key difference is that Preferred stock usually doesn't give shareholders voting rights. Common or ordinary stock does, usually at one vote per share owned.

>Finally, preferred stock is less volatile than common stock and it's callable. 

Callability here just means that **shares can be redeemed by the issuer** i.e. the company issuing the stock can purchase it back at a redemption rate with a premium above the purchase price.

### Bonds
This is possibly the **most important concept to take away**, possibly from the entire workshop series. A solid understanding of how bonds work is crucial to succeeding in the world of finance, even if you will not be working directly in the field. Stocks are straightforward but the world of finance runs much more on bonds than stocks so there is value in taking the time to understand them.

>Bonds are debt instruments and represent loans made to the issuer. Bonds allow individual investors to assume the role of the lender. Governments and corporations commonly use bonds to borrow money to fund roads, schools, dams, or other infrastructure projects.

Essentially bonds can be thought of issuing the ability to 'buy' someone's debt. It serves as an important and fundamental way to raise money for both corporations and governments.

>Bonds are fixed-income securities and are one of the main asset classes for individual investors, along with equities and cash equivalents. The borrower issues a bond that includes the terms of the loan, interest payments that will be made, and the maturity date the bond principal must be repaid. The interest payment is part of the return that bondholders earn for loaning their funds to the issuer. The interest rate that determines the payment is called the coupon rate.

Key terms to note here are **fixed-income security** and **coupon rate**.

#### Fixed-Income Security
>A fixed-income security is an investment that provides a return through fixed periodic interest payments and the eventual return of principal at maturity.

The central idea is that returns are predictable and less volatile (and possibly 'safer' based on the issuer). The tradeoff is that they provide lower returns.

#### Coupon Rate 
The coupon rate is the fixed annual interest rate paid on a bond, expressed as a percentage of the bond's face value. It determines the amount of interest income an investor receives regularly, typically semi-annually or annually, until the bond matures. 



#### Face Value/Par Value
Another key term often brought up is the face value or the par value of a bond. It is essentially just the value of the bond at maturity and what is used to calculate interest payments by the issuer.

There are four primary categories of bonds sold in the markets:

- Corporate Bonds: Companies issue corporate bonds rather than seek bank loans for debt financing because bond markets offer more favorable terms and lower interest rates.
- Municipal Bonds: Issued by states and municipalities. Some municipal bonds offer tax-free coupon income for investors.
- Government Bonds: Bonds issued by the U.S. Treasury with a year or less to maturity are called “Bills,” bonds issued with one–10 years to maturity are called “notes,” and government bonds issued with more than 10 years to maturity are called “bonds.” The entire category of bonds issued by a government treasury is often collectively referred to as "treasuries."
- Agency Bonds: Issued by government-affiliated organizations such as Fannie Mae or Freddie Mac are considered agency bonds.

**Note:** Based on their length of maturity, U.S. government bonds are given different names. This can be confusing however as they may also be just called 'treasuries', 'bonds', or 'government debt' interchangeably meaning more or less the overarching issuance of bonds by the government. Use context to figure out if the specific debt instrument is being talked about. 

### Effect of Interest Rates on Bond Prices
When you own a bond, you have the benefit of being able to sell it freely in the market (assuming ideal conditions). Given the frequency of bond issuance, a key factor to notice is the impact of newly issued bonds on prices of all bonds previously issued.

This is a pretty intuitive relationship: if I held a 4% a year for 10 years bond and the U.S. government decided to come out and issue 5% a year for 10 years bonds, my 4%/10 yr bond is suddenly very unattractive! Thus, the price I would have been able to sell my 4%/10 yr bond will drop to reflect the issuance of rising interest rates. 

Therefore we see **as interest rates rise, bond prices fall, and vice versa**.

2. The slope of the yield curve (the shape) predicts the direction of interest rates (of government bonds) and tells a story about the economic health of the country.

    - Normal Yield Curve: A normal yield curve shows low yields for shorter-maturity bonds increasing for bonds with a longer maturity. The curve slopes upward. This indicates that yields on longer-term bonds continue to rise, responding to periods of economic expansion.
  
      
  
      It shows investors have faith in the long term solvency potential of the government and expect that while they demand higher interest rates on longer maturity bonds, they are comfortable with being paid in 10, 20, 30 years etc.

It makes it more attractive to buy bonds now, increasing demand for long-term securities which offer higher yields for longer (in Normal Yield Curve) than short-term securities.

      Higher bond yields can also be a result of higher inflation. This brings us to the distinction between Real Return vs Nominal Return, where because of higher inflation, people will only demand for bonds if it brings them adequate returns accounting for inflation (if inflation is 3% and a bond returns 4% a year, you are only making 1% real returns which would be unnatractive given other investment options).

   - Inverted Yield Curve: An inverted yield curve slopes downward with short-term interest rates exceeding long-term rates. This type of curve corresponds to a period of economic recession when investors expect yields on longer-maturity bonds to trend lower in the future
  
     

     This is especially the case where investors would prefer being paid out quickly on their bond investments as they might be unsure of how likely they are to be paid out. In times of recession bond yields often fall too, through factors such as a flight to safety from investors to 'safe' securities i.e. government bonds, reduced demand for loans, etc.

### Derivatives

>The term “derivative” refers to a type of financial contract whose value is dependent on an underlying asset, a group of assets, or a benchmark. Derivatives are agreements set between two or more parties that can be traded on an exchange or over the counter (OTC). Prices for derivatives derive from fluctuations in the prices of underlying assets.

These serve some key benefits:
- These contracts can control much more than just 'one' of the asset
- Derivatives give access to markets not open to retail investors i.e. corn, wheat, gold, forex etc. 
- Can be used to hedge risk

**Note:** Over the counter essentially just implies the trading of financial instruments directly between two parties, without the oversight of a centralized exchange like the NYSE or NASDAQ.

Below are the four main types of derivatives and how they are commonly used:

#### Futures Contracts

A futures contract is a standardized agreement to buy or sell an asset at a predetermined price on a specific future date. These contracts trade on organized exchanges like the CME Group and are marked to market daily.

Uses:
- Hedging: A farmer locks in a future price for wheat to protect against a price drop.
- Speculation: A trader bets on oil prices rising by buying oil futures without owning physical oil.

Futures are commonly used for commodities, but also exist for stock indices, interest rates, and foreign currencies.

#### Forward Contracts
A forward contract is similar to a futures contract but is customized, non-standardized, and traded over the counter (OTC). Forwards are settled at the end of the contract and are not marked to market daily.

Uses:
- Typically used by large institutions and corporations to hedge currency risk or commodity prices in private agreements.
- For example, an importer can fix the exchange rate for a future payment in euros to protect against currency fluctuations.

Because they are private contracts, forwards carry counterparty risk—the risk that one party defaults.

#### Options
An option gives the buyer the right (but not the obligation) to buy or sell an asset at a predetermined price before or at expiration.

- **A call option gives the right to buy.**
- **A put option gives the right to sell.**

Uses:
- Hedging: An investor buys a put option to protect against a stock price decline.
- Speculation: A trader buys a call option on Tesla stock expecting its price to rise sharply.
- Income generation: Investors sell covered calls to earn premium income.

Options are widely used for equities, indices, and currencies, and their non-linear payoff makes them a flexible tool for managing risk. **They will be discussed much more extensively in class.**

### Option Premium

The Price paid for the option (without this there would be no benefit for taking the other side of this contract i.e. imagine selling a long call on a stock- if it goes up you have to take the loss and if it stays the same or goes down the buyer does not exercise their right to buy- both scenarios combined, by selling the long call you either lose money or don't make money)

### Strike Price (often referred to as K)

The price at which the options contract can be executed (underlying asset at the price).

### Call Options

A call is an agreement that gives an investor the right, but not the obligation to buy a stock or any other security at the strike price on or before the expiration date. The risk to the call option buyer is limited to the premium paid for the option, which doesn’t include the price of the underlying asset.

### Put Options

A put option is an option contract giving the owner the right, but not the obligation to sell a specified amount of an underlying security at the strike price on or before expiration. The buyer believes the stock’s market price will fall below the strike price (hence buying the ‘right’ to SELL it at a set price in the future). The put option is profitable when the underlying stock’s price is below the strike price. If the prevailing market price is less than the strike price at expiry, the investor can exercise the put by selling shares at the option’s higher strike price. The profit on this trade is the strike price less the current market price, less the premium and any brokerage commission.

#### Swaps
Swaps are agreements where two parties exchange cash flows based on different financial instruments. The most common types are interest rate swaps and currency swaps.

Uses:
- In an interest rate swap, a company paying a variable rate might exchange payments with a company paying a fixed rate to reduce interest rate risk.
- In a currency swap, two firms operating in different countries might exchange payments in different currencies to hedge exchange rate risk.

Swaps are typically used by large corporations, banks, and institutional investors, and are traded OTC.

## Key Market Infrastructure Terms

Here are some terms you will hear often defined. These pertain to the actual functioning of markets and can apply to different securities like stocks, bonds and other financial instruments.

#### Exchanges
Centralized platforms like NYSE or NASDAQ match buyers and sellers. Some assets trade OTC (over the counter), like many bonds.

#### Liquidity
How easily you can buy or sell an asset without affecting its price.

Example: Apple stock is liquid—you can sell 1,000 shares in seconds. A rare painting is illiquid.

#### Volume
How much of a security is available to buy at a given moment of time. For popular securities such as say Apple stock, this may be millions of shares.

#### Bid-Ask Spread
The difference between the bid (buy) and ask (sell) prices.

Example:
- Bid: \\$99.50
- Ask: \\$100.00

Spread: \\$0.50 → this is a hidden transaction cost that an entity like a market maker uses to generate a profit.

#### Order Types

- Market Order: Buy/sell immediately at best available price
- Limit Order: Buy/sell only at a specific price
- Stop Order: Triggered at a price threshold (used for protection)

### Efficient Markets Hypothesis

>The efficient market hypothesis (EMH), alternatively known as the efficient market theory, is a hypothesis that states that share prices reflect all available information and consistent alpha generation is impossible.

Here **alpha** means a profit above what the market expects from the price of the security to be. Say everyone expected the price of an apple to be \\$3 and you found a way to buy apples for \\$1 and sell them at the market for the aforementioned \\$2 you would have made an alpha of \\$2.

EMH exists primarily in three forms based on how true one assumes it to be:

- Weak Form: Prices reflect past prices — charting won’t help
- Semi-Strong Form: Prices reflect all public info — earnings news priced in instantly
- Strong Form: Even insider info is priced in — no one can beat the market

The key takeaway here should be that economists pose through the EMH that Beating the market consistently is nearly impossible. There is always the idea of **mean-reversion** i.e. the natural tendency of securities that are over/under performing to revert to their mean/expected values.

### Capital Asset Pricing Model (CAPM)



This will be discussed **very extensively**, especially in BUFN402, however I just wanted to introduce the concept here. It is essentially the relationship between systematic risk, or the general perils of investing, and expected return for assets, particularly stocks.

> The goal of the CAPM formula is to evaluate whether a stock is fairly valued when its risk and the time value of money are compared with its expected return. In other words, by knowing the individual parts of the CAPM, it is possible to gauge whether the current price of a stock is consistent with its likely return.

If the EMH is held to be untrue, there is also a value of alpha added to the equation.
