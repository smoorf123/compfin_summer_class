### Monetary Policy

>Monetary policy refers to the actions of central banks, including the Federal Reserve, to achieve macroeconomic policy objectives such as price stability, full employment, and stable economic growth.

Here we come across a couple of crucial concepts, such as the **Federal Reserve as a Central Bank, Price Stability, Full Employment, and Stable Economic Growth.** Lets define these terms and then go back to our definition. Firstly, **what is a central bank and why is it important?**

>A central bank is a financial institution given privileged control over the production and distribution of money and credit for a nation, union, or group of countries. In modern economies, the central bank is usually responsible for formulating monetary policy and regulating member banks.
>
>The Fed has a central governing board of seven members and 12 regional Federal Reserve banks, each responsible for a specific geographic area of the U.S., with its own president.

Whats important to note is the Fed's relationship to the government. It is generally considered an independent entitiy (non-partisan) only reporting to Congress, however, the Federal Reserve Banks that it oversees are set up like private corporations and are not directly accountable to the government.

Now that we know what a central bank is (the Fed, in this case), we can discuss how the Fed achieves their policy goals and what their overarching mandate is. Generally we call the Fed's influence on the economy as **Monetary Policy**.

>The U.S. Congress established maximum employment and price stability as the macroeconomic objectives for the Fed; they are referred to as the Fed's dual mandate.

The term **dual mandate** will come up often. It colloquially refers to the two main goals of the Fed namely maximizing employment and price stability. 

1) Maximizing Employment: Maximum (or full) employment is the total measure of employment that the economy can experience without any overt inflationary pressures i.e. too much employment causing pressure on prices of goods. Almost everyone who wants a job can secure one during maximum employment though the goal is not to reach 100% employment and completely eradicate unemployment.

2) Price Stability: People and businesses need to be reasonably confident that prices will remain relatively constant over time so they can plan for the future. As a result, unstable prices (through either deflation or rapid inflation) can have drastic consequences on economic stability.

**Key Takeaway:** It is impossible to achieve both goals at once, since low unemployment leads to high prices, and vice versa. Instead, the Fed is continually adjusting interest rates to balance the two priorities. It aims to keep unemployment low, but not zero. It also aims to **limit price growth (i.e. inflation) at about 2% per year.**

The tool the Fed primarily has to achieve these goals is through influencing the **(effective) Federal Funds Rate**, which is the target interest rate at which commercial banks borrow and lend their excess reserves to each other overnight, set by the **Federal Open Market Committee (FOMC)** (more on them later). This is generally a range (e.g. 4.25% - 4.5%) and is modified in incremements of 25 basis points (0.25%) in normal market conditions or 50 basis points (0.5%) in times of uncertainity.

This functionality is serviced through legal liquidity requirements.

>The law requires that banks must have a minimum reserve level in proportion to their deposits. This reserve requirement is held at a Federal Reserve bank. When a bank has excess reserve requirements, it may lend these funds overnight to other banks that have realized a reserve deficit.

The rate at which banks lend to each other is crucial as it determines a wide range of economic activity such as:

- **(Important Factor)** Real estate: Mortgage rates often move in tandem with the federal funds rate, affecting home buying and refinancing activity. It's a major reason the Fed Funds rate makes national news headlines.
- Automotive industry: Car loan rates typically follow trends in the federal funds rate, impacting vehicle sales.
- Consumer lending: Credit card rates and personal loan rates are influenced by changes in the federal funds rate.
- Foreign exchange: Rate changes can affect the value of the U.S. dollar relative to other currencies, impacting international trade and investment.
- Government finance: The rate influences the cost of government borrowing and can affect budget deficits.
- Financial markets: Changes in the federal funds rate can affect stock prices since it affects corporate profits and investor behavior.

**The Federal Open Market Committee (FOMC)** is critical to the Fed's operations. 

> The Federal Open Market Committee (FOMC), which meets eight times per year to discuss and review economic and financial developments, determines the appropriate stance of monetary policy. It is customary for the Federal Open Market Committee to meet eight times annually to determine the federal funds rate. These rates are influenced by economic indicators, such as the core inflation rate and the durable goods orders report, which provide signals about the economic health of the country.

Now that we know what the FOMC is and their task of determining the target range for the Fed Funds rate, we can discuss how the Fed influences the rate at which Banks lend to each other. This is no easy task and requires careful planning and large-scale buy-in. (Why?)

To affect the Fed Funds rate The Fed engages in something called **Open Market Operations (OMO)**.

>The U.S. central bank can lower the interest rate by purchasing securities (and injecting money into the money supply). Similarly, it can sell securities from its balance sheet, take money out of circulation, and put upward pressure on interest rates.

Open Market Operations are complex and could take up a whole class or possibly multiple classes. In the interest of being succinct, I will leave them as optional further readings if you are interested. This would include understanding Permanent vs Temporary OMOs, Expansionary policy vs Contractionary policy, Quantatative Easing vs Quantatative Tightening, Repo agreements, Reverse Repo agreements, etc. 

These topics help illustrate the complexity of the job the Fed has (and why they hire hundreds of economists) as well as the fact maintaining the economy is far from an exact science. There is always new research being produced and conflicting opinions (try watching CNBC close to a FOMC meeting, especially if the economy is under fire). **Bonus Content:** If you keep up with financial news you might have heard about the "Fed Dot Plot". The Fed publishes its projections four times a year; the “dot plot” can be found in the pages of its Summary of Economic Projections reports, allowing you to track how each FOMC member thinks interest rates will be by the end of the current year, two or three (depending on the time of year) consecutive years after, and the more ambiguous “longer run.” Each “dot” represents a member’s individual view, however identities are kept anonymous.

Can you use the dot plot to predict future rate decisions? Sort of, but not reliably. That’s because the Fed members will likely shift their projections as new economic data rolls in. But you can use it to better understand what the Fed is currently thinking and what it might do in the next FOMC meeting.



If you compare dot plots over time, you can get a sense of whether the Fed is leaning hawkish (and looking to raise interest rates) or dovish (and looking to possibly lower rates). Fed hawkishness suggests that it’s concerned about inflation.

### Fiscal Policy

>Fiscal policy refers to the tax and spending policies of a national government. In the U.S., fiscal policy decisions are determined by Congress and the Administration; the Fed plays no role in determining fiscal policy.

Fiscal policy refers to the steps that governments take in order to influence the direction of the economy. But rather than encouraging or restricting spending by businesses and consumers, fiscal policy aims to target the total level of spending, the total composition of spending, or both in an economy.

The two most widely used means of affecting fiscal policy are:

1) Government Spending Policies: Governments can increase the amount of money they spend if they believe there is not enough business activity in an economy. This is often referred to as stimulus spending. They can borrow money by issuing debt securities (like government bonds) if there are not enough tax receipts to pay for the spending increases, allowing them to accumulate debt. This is referred to as deficit spending.

2) Government Tax Policies: By increasing taxes, governments pull money out of the economy and slow business activity. Fiscal policy is typically used when the government seeks to stimulate the economy. It might lower taxes or offer tax rebates in an effort to encourage economic growth.
 
When a government spends money or changes tax policy, it must choose where to spend or what to tax. In doing so, government fiscal policy can target specific communities, industries, investments, or commodities to either favor or discourage production—sometimes, its actions are based on considerations that are not entirely economic. For this reason, fiscal policy is often hotly debated among economists and political observers.

While Fiscal policy is a joint effort between the legislative and executive branches of government, it is important to note the role of the **Treasury**.

>The primary task of the Treasury secretary is to advise the president on domestic and international economic issues and implement the administration's economic policies.
>
>The Department provides economic advice to the president and works with other federal institutions, including the Federal Reserve, to "encourage global economic growth, raise standards of living, and to the extent possible, predict and prevent economic crises."

The responsibilities of the Treasury are expansive, including but not limited to:

- Internal Revenue Service (IRS)
- Bureau of the Fiscal Service
- U.S. Mint
- Financial Crimes Enforcement Network
- Office of the Comptroller of the Currency (OCC)
- Alcohol and Tobacco Tax and Trade Bureau
 
The Treasury really is a treasury, too. It stores most of the nation's gold supply in a vault at the New York Fed. This is one example of how the responsibilities of the Treasury and the Federal Reserve overlap

To note is the distinction between the Fed and the Treasury, as well as when their roles overlap.

1. The Federal Reserve conducts Treasury securities auctions on behalf of the Department of the Treasury. Examples of Treasury securities include (treasury securities are divided into three primary categories according to the length of maturity):

- Treasury bonds
- Treasury bills
- Treasury notes

Treasury inflation-protected securities (TIPS) are also auctioned however will not be discussed here. Feel free to look up for further reading.

We will cover more specifics about Treasuries in later lectures (Investment Products Lectures) however for a basic overview, 

>Government securities are government debt issuances used to fund daily operations, and special infrastructure and military projects. They guarantee the full repayment of invested principal at the maturity of the security and often pay periodic coupon or interest payments. Government securities are considered to be risk-free as they have the backing of the government that issued them. The tradeoff of buying risk-free securities is that they tend to pay a lower rate of interest than corporate bonds.

*The question of whether these securities are truly "risk-free" may be up for debate and is an important nuance to consider, especially when talking about non-US debt issuance (a.k.a non-US treasuries).*

2. The Treasury is responsible for managing revenues and paying government debts, while the Fed acts as the nation's central bank.

3. The Federal Reserve is a nonprofit entity. Any remaining profits are paid to the Department of the Treasury after its expenses are paid. The Department of the Treasury then uses that money to fund government spending.

It's a relationship that produces a considerable amount of money. The Federal Reserve contributed an estimated $79.1 billion to the Treasury in 2024.

**There are many other distinctions and collaborations given the complexity of the task of maintaining the economy however the fundamental takeaway, besides the key difference in roles, should be the intentional non-partisanship of the Fed as opposed to the Treasury.** In 1951, the two bodies enacted the Treasury-Fed accord to ensure the independence of the Fed. This agreement was intended to protect monetary policy from political manipulation.

This explicit distinction becomes increasingly apparent in times of high political divide, such as these.
